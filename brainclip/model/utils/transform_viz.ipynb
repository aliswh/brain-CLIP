```python
from brainclip.model.network.data_loader import BrainCLIPDataLoader
from brainclip.model.utils.transforms import apply_transform, get_transforms
```

```python
train_loader = BrainCLIPDataLoader("train", batch_size=3)
val_loader = BrainCLIPDataLoader("valid", batch_size=3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torchio as tio
import torch
import nibabel as nib
```

```python
import matplotlib.pyplot as plt

def plot_batch(batch):
    # Get the first sample in the batch
    sample = batch[0]
    
    # Get the first channel and the 10th slice of the z-dimension
    slice_ = sample[0, :, :, 8]
    
    # Plot the slice
    plt.imshow(slice_, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
i=0
for image, input_id_report, attention_mask_report, label, image_path in train_loader:
    i+=1
    plot_batch(image)
    if i==5:break
```

```python
i=0
for image, input_id_report, attention_mask_report, label, image_path in val_loader:
    i+=1
    plot_batch(image)
    if i==5:break
```
