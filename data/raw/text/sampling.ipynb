```python
import pandas as pd
```

```python
nlp_file = "/home/alice/Desktop/NLP/nlp_diagnosis_revised.xlsx"
df = pd.read_excel(nlp_file)
classes = ["infarct","tumor","hemorrhage","others","normal"]
```

```python
print(len(df))
for row in range(len(df)):
    if sum(df[classes].loc[row]) > 1:
        df = df.drop(row, axis=0)
len(df)
```

```python
classes_col = df[classes].idxmax(axis=1)
df["class"] = classes_col
df
```

```python
n = 30
d = {c:None for c in classes}
id_cols = ["Dataset","StudyInstanceUID","ParsedImpressions","class"]	

for c in classes:
    print(c, len(df[df["class"] == c]))
    d[c] = df[df["class"] == c].sample(n, random_state=42)[id_cols] 
```

```python
e = pd.concat(list(d.values()))
e
```

```python
n = 10
dd = {c:None for c in classes}
id_cols = ["Dataset","StudyInstanceUID","ParsedImpressions", "class"]	

for c in classes:
    dff = d[c]
    dd[c] = dff[dff["class"] == c].sample(n, random_state=42)
```

```python
r = pd.concat(list(dd.values()))
r.to_csv("/home/alice/Desktop/NLP/sampled_docs.csv")
r
```
