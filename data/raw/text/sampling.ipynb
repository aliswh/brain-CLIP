```python
import pandas as pd
```

```python
nlp_file = "/datadrive_m2/alice/brain-CLIP/data/raw/text/nlp_diagnosis_revised.xlsx"
df = pd.read_excel(nlp_file)
df.columns = ["ID", *df.columns[1:]]
classes = ["infarct","tumor","hemorrhage","others","normal"]
```

```python
print(len(df))
for row in range(len(df)):
    if sum(df[classes].loc[row]) > 1:
        df = df.drop(row, axis=0)
len(df)
```

```python
classes_col = df[classes].idxmax(axis=1)
df["class"] = classes_col
df
```

```python
list(df[df["ID"]==1027]["ParsedImpressions"])
```

```python
for c in classes: print(c, len(df[df["class"] == c]))
```

```python
n = 60
dd = {c:None for c in classes}
id_cols = ["ID","Dataset","StudyInstanceUID","ParsedImpressions", "class"]	

for c in classes:
    df_class_c = df[df["class"] == c]
    num_samples = min(len(df_class_c), n)
    print(c, num_samples)
    dd[c] = df_class_c.sample(num_samples, random_state=42)
```

```python
r = pd.concat(list(dd.values()))
r
```
